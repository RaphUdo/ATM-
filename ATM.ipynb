```python
bank_balance = 1000
def display_balance():
  print(f"Your current balance is £{bank_balance}")
def withdraw_money():
  global bank_balance
  amount = float(input("Enter the amount to withdraw: £"))
  if amount > bank_balance:
    print("Insufficient funds. Withdrawal Cancled.")
  else:
    bank_balance -= amount
  print(f"Withdrew £{amount}. Remaining: £{bank_balance}")

def main_menu():
  print("\nATM Menu:")
  print("1. Display Balance")
  print('2. Withdraw Money')
  print('3. Exit')
  choice = input("Enter your choice")
  if choice == "1":
    display_balance()
  elif choice == "2":
    withdraw_money()
  elif choice == "3":
    print("Thank you for using this ATM. Goodbye!")
  else:
    print("Invalid choice. Please select a valid option.")
```

```python
main_menu()
```
